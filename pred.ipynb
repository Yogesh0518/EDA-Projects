```python
def polynomial_equation(X):
    a0 = 0.959553035936221
    a1 = 3.2273475913782166
    a2 = 0.08499756750243252
    a3 = -0.005200288784683211
    a4 = 0.006832803338671797
    a5 = 1.1881667503666924e-05
    a6 = -0.0019633456608741523
    a7 = -0.003220532340222836
    a8 = -1.7774496751611928e-05

    x1 = X['20m Avg [m/s]']
    x2 = X['Temp 5m [°C]']
    x3 = X['Pressure 5m [mbar]']

    # Compute the polynomial equation
    y = inter + a0*x1 + a1*x2 + a2*x3 + a3*(x1**2) + a4*(x2**2) + a5*(x3**2) + a6*(x1*x2) + a7*(x2*x3) + a8*(x1*x3)
    return y
```
