# RANDOM FOREST

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from sklearn.model_selection import train_test_split,GridSearchCV
def mean_absolute_percentage_error(y_test1, y_predc1):
    y_test1, y_predc1 = np.array(y_test1), np.array(y_predc1)
    return np.mean(np.abs((y_test1 - y_predc1) / y_test1)) * 100
```

```python
X = reg_data[["Year","Month","Date"]]  # Replace with the appropriate column names
Heat_pump = reg_data["Heat_pump"]  # Replace 'target' with the appropriate column name
X_train, X_test, y_train, y_test = train_test_split(X, Heat_pump, test_size=0.33333, random_state=42)

# Create and fit the random forest regression model
model = RandomForestRegressor()
model.fit(X_train, y_train)
Heat_pump_pred_rf = model.predict(X_test)

#ACCURACY CHECKING
mse = mean_squared_error(y_test, Heat_pump_pred_rf)
print("RMSE:",mse**(1/2.0)) 
rfmae = mean_absolute_error(y_test, Heat_pump_pred_rf)
print("MAE: ",rfmae)
mape = mean_absolute_percentage_error(y_test, Heat_pump_pred_rf)
print(f"MAPE: {mape:.2f}%")
r2=r2_score(y_test,Heat_pump_pred_rf)
print("R2:",r2)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from xgboost import XGBRegressor, DMatrix
from sklearn.preprocessing import OneHotEncoder

```

```python
X = reg_data[["Year","Month","Date"]] # Replace 'target_variable_column' with the name of your target variable column
Heat_pump = reg_data["Heat_pump"]
X_train, X_test, y_train, y_test = train_test_split(X, Heat_pump, test_size=0.33333, random_state=42)
dtrain = xgb.DMatrix(X_train,label=y_train)
dtest = xgb.DMatrix(X_test,label=y_test)
params = {
   'objective': 'reg:squarederror',  # Use 'reg:squarederror' for regression
   'eval_metric': 'rmse'  # Root Mean Squared Error (RMSE) as the evaluation metric
}
num_rounds = 100  # Adjust the number of boosting rounds
model = xgb.train(params,dtrain,num_rounds)
pump_pred_xgb = model.predict(dtest)
pump_pred_xgb

#ACCURACY CHECK
mse = mean_squared_error(y_test, pump_pred_xgb)
print("MSE: ",mse)
print("RMSE: ",mse**(1/2.0))
mape = mean_absolute_percentage_error(y_test,pump_pred_xgb)
print(f"MAPE: {mape:.2f}%")
r2=r2_score(y_test,pump_pred_xgb)
print("R2:",r2)
```

#### DEGREE 1

```python
import matplotlib.pyplot as plt

plt.scatter(X,y, color="red")
plt.plot(X, lin_reg.predict(X))
plt.title("Linear Regression")
plt.xlabel("20m")
plt.ylabel("100m")
plt.show()
```

```python
lin_reg.predict([[4.068]])
```

#### DEGREE 2

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("Poly Regression - Degree 2")
plt.xlabel("20m")
plt.ylabel("100m")
plt.show()
```

```python
new_salary_pred = lin_reg_2.predict(poly_reg.fit_transform([[4.068]]))
print('The predicted salary of a person at 6.5 Level is ',new_salary_pred)
```

#### DEGREE 3

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)

 # Step 6 - Passing X_poly to LinearRegression
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)

# Step 7 - Visualize Poly Regression Results
plt.scatter(X,y, color="red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("Poly Regression Degree 3")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# Step 8 Polynomial Regression prediction
new_salary_pred = lin_reg_2.predict(poly_reg.fit_transform([[4.068]]))
```
